# Réalisez une étude de santé publique sur l'année 2017

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Aide alimentaire

```python
#Import du csv aide alimentaire
#Affichage du type de donnée contenue dans le dataframe

aide_alim = pd.read_csv('aide_alimentaire.csv')
display(aide_alim.head(10))
aide_alim.dtypes
```

```python
# Remplacement des NaN

aide_alim.fillna(0, inplace=True)
aide_alim.head(10)
```

### Dispo_Alimentaire

```python
#Import du csv disponibilité alimentaire
#Affichage du type de donnée contenue dans le dataframe

dispo_alim = pd.read_csv('dispo_alimentaire.csv')
display(dispo_alim.head(10))
dispo_alim.dtypes
```

```python
# Remplacement des NaN

dispo_alim.fillna(0, inplace=True)
dispo_alim.head(10)
```

### Population

```python
#Import du csv population
#Affichage du type de donnée contenue dans le dataframe

pop = pd.read_csv('population.csv')
display(pop.head(10))
pop.dtypes
```

```python
# Nom de la colonne valeur remplacée par Population 

pop = pop.rename(columns={'Valeur': 'Population'})
display(pop.head(10))
pop.dtypes
```

```python
# Remplacement des NaN

pop.fillna(0, inplace=True)
pop.head(10)
```

```python
# Conversion de l'unité de Population

pop['Population']= pop['Population']* 1000
pop.head()
```

### Sous nutrition

```python
#Import du csv sous nutrition
#Affichage du type de donnée contenue dans le dataframe

sous_nut = pd.read_csv('sous_nutrition.csv')
display(sous_nut.head(70))
sous_nut.dtypes
```

```python
# La colonne valeur est en objet car il y a des Nan et <0.1
# Remplacement des <0,1 par 0 
# Exemple avec la ligne 65 Arménie <0.1 ainsi que les lignes Australie NaN
# On renomme aussi la colonne pour être plus explicite

sous_nut.rename(columns={"Valeur": "Sous_nutrition"}, inplace=True)
sous_nut.fillna(0, inplace=True)
sous_nut['Sous_nutrition'] = pd.to_numeric(sous_nut['Sous_nutrition'], errors='coerce')
display(sous_nut.head(70))
sous_nut.dtypes
```

```python
# Pour simplifier on multiplie par 1 000 000 pour avoir en millions d'habitants et on renomme la colonne

sous_nut['Sous_nutrition'] *= 1000000
sous_nut.head(70)
```

## 1 . La proportion de personnes en état de sous-nutrition en 2017


```python
# Il faut tout d'abord faire une jointure entre population et sous nutrition, en ciblant l'année 2017 :
prop_sous_nut = pd.merge(pop.loc[pop['Année'] == 2017,["Zone", "Population"]],
                               sous_nut.loc[sous_nut['Année'] == '2016-2018',["Zone", "Sous_nutrition"]],
                               on='Zone')
prop_sous_nut.head(10)
```

```python
print("Proportion de personnes en état de sous nutrition :", "{:.2f}".format(prop_sous_nut['Sous_nutrition'].sum()*100/prop_sous_nut['Population'].sum()), "%")
```

## 2. Le nombre théorique de personnes qui pourraient être nourries avec la disponibilité alimentaire mondiale en 2017


```python
# Il faut tout d'abord faire une jointure entre population et disponibilité alimentaire, en ciblant l'année 2017 :

dispo_alimentaire2017 = dispo_alim.merge(pop.loc[pop['Année'] == 2017,["Zone", "Population"]],
                                            on='Zone')
display(dispo_alimentaire2017.head(10))


```

```python
#Disponibilté alimentaire totale en kcalories

dispo_alimentaire2017['dispo_kcal'] = dispo_alimentaire2017['Disponibilité alimentaire (Kcal/personne/jour)'] * dispo_alimentaire2017['Population'] * 365
print("dispo alimentaire totale en kcal :", dispo_alimentaire2017['dispo_kcal'].sum(), "kcal")
```

```python
#Calcul du nombre d'humains pouvant être nourris
#Besoin kcal moyen adulte = 2500 kcal/jour (Selon l'ONU, on considère les besoins d'un être humain à environ 2500 kcal/jour.)

energie_kcal = dispo_alimentaire2017['dispo_kcal'].sum()/(2500*365)
print("Total d'êtres humains pouvant être nourris :", energie_kcal)
print("Proportion de la population :", "{:.2f}".format(energie_kcal*100/pop.loc[pop['Année'] == 2017,"Population"].sum()), "%")
```

## 3. Le nombre théorique de personnes qui pourraient être nourries avec la disponibilité alimentaire des produits végétaux

```python

dispo_vegetale2017 = dispo_alim.merge(pop.loc[pop['Année'] == 2017,["Zone", "Population"]],
                                            on='Zone')
display(dispo_vegetale2017.head(10))

```

```python
#On renomme zone en Pays 
dispo_vegetale2017.rename(columns={'Zone': 'Pays'}, inplace= True)
dispo_vegetale2017
```

```python
dispo_vegetale2017 = dispo_vegetale2017.copy()
dispo_vegetale2017['dispo_kcal_vegetale'] = dispo_vegetale2017['Disponibilité alimentaire (Kcal/personne/jour)'] * dispo_vegetale2017['Population'] * 365
dispo_vegetale2017
print("dispo alimentaire totale en kcal uniquement végétale :", dispo_vegetale2017['dispo_kcal_vegetale'].sum(), "kcal")
```

```python
#Calcul du nombre d'humains pouvant être nourris
#Besoin kcal moyen adulte = 2500 kcal/jour (Selon l'ONU, on considère les besoins d'un être humain à environ 2500 kcal/jour.)

energie_kcal = dispo_vegetale2017['dispo_kcal_vegetale'].sum()/(2500*365)
print("Total d'êtres humains pouvant être nourris :", energie_kcal)
print("Proportion de la population :", "{:.2f}".format(energie_kcal*100/pop.loc[pop['Année'] == 2017,"Population"].sum()), "%")
```

## 4. L’utilisation de la disponibilité intérieure:
### - la part qui est attribuée à l’alimentation animale, 
### - celle qui est perdue 
### - et celle qui est concrètement utilisée pour l'alimentation humaine

```python
# Disponibilité Interieure = (Production + Variation de stock + Importations) - Exportations
```

 - la part attribuée à l’alimentation animale

```python
# Il faut tout d'abord faire une jointure entre population et disponibilité alimentaire, en ciblant l'année 2017 :

dispo_2017 = dispo_alim.merge(pop.loc[pop['Année'] == 2017,["Zone", "Population"]],
                                            on='Zone')
dispo_2017.head()
```

```python
#Dispo interieure totale

dispo_int = dispo_2017 ['Disponibilité intérieure'].sum()
dispo_int
```

```python
dispo_anim = dispo_2017 ['Aliments pour animaux'].sum()
dispo_anim

```

- celle qui est perdue

```python
dispo_perdue = dispo_2017 ['Pertes'].sum()
dispo_perdue
```

- celle qui est concrètement utilisée pour l'alimentation humaine

```python
dispo_hum = dispo_2017 ['Nourriture'].sum()
dispo_hum
```

```python
# Ratio Alimentation animale / Disponibilité interieure
dispo_2017 ['Alim_anim/dispo_interieure_%'] = (dispo_anim / dispo_int) * 100

# ratio "Pertes" / "Disponibilité interieure"
dispo_2017 ['Pertes/dispo_interieure_%'] = (dispo_perdue / dispo_int) * 100

# Ratio Alimentation humaine / Disponibilité interieure
dispo_2017 ['Alim_hum/dispo_interieure_%'] = (dispo_hum / dispo_int) * 100

dispo_2017.head()
```

```python
labels = 'Alim_anim/dispo_interieure_%', 'Pertes/dispo_interieure_%', 'Alim_hum/dispo_interieure_%'
sizes = [13.23, 4.65, 49.37]

plt.title('Répartition des ratio de disponibilité interieure')
plt.pie(sizes,labels = labels)
plt.show()
```

```python
# Infos en plus , ratio par pays

# Ratio Alimentation animale / Disponibilité interieure
dispo_2017 ['Alim_anim/dispo_interieure_%'] = (dispo_2017 ['Aliments pour animaux'] / dispo_2017 ['Disponibilité intérieure']) * 100

# ratio "Pertes" / "Disponibilité interieure"
dispo_2017 ['Pertes/dispo_interieure_%'] = (dispo_2017 ['Pertes'] /dispo_2017 ['Disponibilité intérieure']) * 100

# Ratio Alimentation humaine / Disponibilité interieure
dispo_2017 ['Alim_hum/dispo_interieure_%'] = (dispo_2017['Nourriture'] /dispo_2017 ['Disponibilité intérieure']) * 100

dispo_2017.head()
```

## 5. Utilisation des céréales, notamment la répartition entre l’alimentation humaine (colonne Nourriture) et l’alimentation pour animaux

```python
# Céréales destinées à l'alimentation

cereales_pr_alim = dispo_alim[(dispo_alim['Aliments pour animaux'] > 0) | (dispo_alim['Nourriture'] > 0)]

cereales_animaux_alim = cereales_pr_alim['Aliments pour animaux'].sum()
cereales_alim = cereales_pr_alim['Nourriture'].sum() + cereales_animaux_alim


print("Total proportion de la nourriture animale dans l'alimentation (céréales seulement) : {:.2f} %".format((cereales_animaux_alim/cereales_alim) * 100))
```

## 6. Utilisation du manioc par la Thaïlande aux égards de la proportion de personnes en sous-nutrition. (l’exportation par rapport à la production)

```python
# Calcul de la proportion de manioc exporté par la Thaïlande

manioc_thailand = dispo_2017[(dispo_2017 ['Zone'] == 'Thaïlande') & (dispo_2017['Produit'] == 'Manioc')].copy()
manioc_thailand[['Disponibilité intérieure',
       'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Pertes',
       'Production', 'Semences', 'Traitement', 'Variation de stock']].head()

# Ratio "Export"/"Total"
manioc_thailand['ratio_export_total_%'] = (manioc_thailand['Exportations - Quantité'] / (manioc_thailand['Production'] + manioc_thailand['Importations - Quantité'] + manioc_thailand['Variation de stock'])) * 100

print("La proportion du Manioc exporté pour la Thailand est {:.2f}% de la quantité totale de Manioc disponible.".format(manioc_thailand['ratio_export_total_%'].sum()))
```
